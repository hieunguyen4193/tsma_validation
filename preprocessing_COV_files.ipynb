```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
import scipy
import pathlib 
from tqdm import tqdm
import os 
from scipy import optimize

covdir = "/Volumes/HNSD01/storage/TSMA_validation_cov"
outdir = "/Volumes/HNSD03/outdir/TSMA_validation"
path_to_main_output = os.path.join(outdir, "analysis_output")
path_to_save_output = os.path.join(path_to_main_output, "processed_cov")
os.system(f"mkdir -p {path_to_save_output}")

paneldf = pd.read_csv("all_cpgs_in_all_panels.csv")
paneldf["cpg"] = paneldf[["chrom", "start", "end"]].apply(lambda x: f"{x[0]}:{x[1]}-{x[2]}", axis = 1 )
all_covfiles = [item for item in pathlib.Path(covdir).glob("*.cov")]

for file in tqdm(all_covfiles):
    sampleid = file.name.split("_")[0].split("-")[1]
    os.system(f"mkdir -p {path_to_save_output}/{sampleid}")

    tmpdf = pd.read_csv(file, sep = "\t", header = None)
    tmpdf.columns = ["chrom", "start", "end", "methyl", "countC", "countT"]
    tmpdf["cpg"] = tmpdf[["chrom", "start", "end"]].apply(lambda x: f"{x[0]}:{x[1]}-{x[2]}", axis = 1 )

    tmpdf_tsma = tmpdf[tmpdf["cpg"].isin(paneldf[paneldf["panel"] == "TSMA_panel"]["cpg"].to_list())]
    tmpdf_rare_cancer = tmpdf[tmpdf["cpg"].isin(paneldf[paneldf["panel"] == "Rare_cancers_panel"]["cpg"].to_list())]
    tmpdf_450 = tmpdf[tmpdf["cpg"].isin(paneldf[paneldf["panel"] == "TMD_450_panel"]["cpg"].to_list())]

    tmpdf_tsma.to_csv(os.path.join(path_to_save_output, sampleid, f"{sampleid}_TSMA.csv"))
    tmpdf_rare_cancer.to_csv(os.path.join(path_to_save_output, sampleid, f"{sampleid}_RareCancer.csv"))
    tmpdf_450.to_csv(os.path.join(path_to_save_output, sampleid, f"{sampleid}_TMD450.csv"))
```
